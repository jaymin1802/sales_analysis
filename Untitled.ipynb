```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path="./Sales_Data"

files=[file for file in os.listdir(path) if not file.startswith('.')]

df=pd.DataFrame()

for file in files:
    current_data=pd.read_csv(path+'/'+file)
    df=pd.concat([current_data,df])

df.to_csv("final_data.csv",index=False)
```

```python
def format(x):
    if x>=10_000_000:
        return f"{x/10_000_000:.2f} CR"
    elif x>= 100_000:
        return f"{x/100_000:.2f} L"
    elif x>=1000:
        return f"{x/1000:.2f} K"
    else:
        return f"{x:.2f}"
```

```python
hour_sale=df.groupby('order_hour')['revenue'].sum().to_frame('total')
```

```python
plt.plot(hour_sale.index,hour_sale['total'],marker='o')
plt.title('HOUR TRADE')
plt.xticks(hour_sale.index)
plt.xlabel('HOUR')
plt.ylabel('REVENUE')
plt.grid(True)
plt.show()
```

```python
top_product=df.groupby('order_city')['revenue'].sum().to_frame('total')
```

```python
top_products=df.groupby('product')['revenue'].sum().to_frame('sales')
top_products=top_products['sales'].apply(format)
top_products
```

```python
plt.barh(top_products.index,top_products)
plt.title('PROUDCT WISE SALES')
plt.xticks(top_products,rotation=90)
plt.xlabel('REVENUE')
plt.ylabel('PRODUCTS')
plt.grid(True)
plt.show()
```

```python
product_wise_quantity=df.groupby('product')['quantity_ordered'].count()
product_wise_price=df.groupby('product')['price_each'].mean()
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(product_wise_quantity.index, product_wise_quantity, color='g')
ax2.plot(product_wise_price.index, product_wise_price, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(product_wise_price.index, rotation='vertical', size=8)
ax1.grid(True)
fig.show()
```
